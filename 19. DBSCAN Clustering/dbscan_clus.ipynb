```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
X = df.iloc[:,-2:].values
X[:5]
```

```python
sns.scatterplot(x= X[:,0],y=X[:,1],data=X)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
k_rng = range(1,11)
for k in k_rng:
    km = KMeans(k)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(k_rng,wcss)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=5,min_samples=5)
```

```python
dbscan
```

```python
dbscan.fit(X)
```

```python
labels = dbscan.labels_
```

```python
labels
```

```python
from sklearn import metrics
```

```python
print(metrics.silhouette_score(X,labels))
```

```python
sns.scatterplot(x = X[labels==-1,0], y = X[labels==-1,1])
sns.scatterplot(x = X[labels==0,0], y = X[labels==0,1])
sns.scatterplot(x = X[labels==1,0], y = X[labels==1,1])
sns.scatterplot(x = X[labels==2,0], y = X[labels==2,1])
sns.scatterplot(x = X[labels==3,0], y = X[labels==3,1])
sns.scatterplot(x = X[labels==4,0], y = X[labels==4,1])
```
