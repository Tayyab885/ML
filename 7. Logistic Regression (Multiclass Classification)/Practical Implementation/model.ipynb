```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
dataset.target
```

```python
X = df
X
```

```python
y = pd.DataFrame(dataset.target,columns=['target'])
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
params = [{'C':[0.1,0.5,1,5,10,50,100,500,1000]},{'max_iter':[100,500,1000]}]
```

```python
model = LogisticRegression(C=100,max_iter=100)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
linear_regression = GridSearchCV(model, params, scoring='f1' ,cv=5)
```

```python
linear_regression.fit(X_train, y_train)
```

```python
print(linear_regression.best_params_)
print(linear_regression.best_score_)
```

```python
y_predict = linear_regression.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cme = confusion_matrix(y_test, y_predict)
cme
```

```python
sns.heatmap(cme, annot=True, fmt='d', cmap='Blues')
```

```python
accuracy_score(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```
